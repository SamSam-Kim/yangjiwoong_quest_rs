# 배경 블러처리 Flow

```python
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

```python
img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/songgang.jpeg' 
# 원본 이미지 저장
img_orig = cv2.imread(img_path)
# RGB형태로 변경
img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)

print(img_orig.shape)
plt.imshow(img_orig)
plt.axis('off')
plt.show()
```

## 세그멘테이션 모델 생성
+ deeplabv3_xception 모델 사용
+ https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5

```python
model_path = os.getenv('HOME')+'/aiffel/human_segmentation/models/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5' 

model = semantic_segmentation() # DeepLab v3+ 모델
model.load_pascalvoc_model(model_path) # 가중치를 채워넣음

segvalues, output = model.segmentAsPascalvoc(img_path)
```

```python
#pascalvoc 데이터의 라벨종류
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
len(LABEL_NAMES)
```

## 세그멘테이션

```python
# 세그멘테이션 결과 시각화
# output이 BGR형식이기때문에 RGB로 변환
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

print(output.shape)
plt.imshow(output)
plt.show()
```

```python
#segvalues에 있는 class_ids를 담겨있는 값을 통해 pacalvoc에 담겨있는 라벨을 출력
for class_id in segvalues['class_ids']:
    print(LABEL_NAMES[class_id])
```

## 사람에 대한 세그멘테이션 마스크 생성

```python
# 클래스 ID별로 지정된 RGB 색상을 담고있는 colormap 생성 코드(이해할 필요 없음. PixelLib에서 그대로 가져온 코드)
colormap = np.zeros((256, 3), dtype=int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3
```

```python
person_seg = colormap[15] #컬러맵 15(사람)에 해당하는 RGB값 저장
```

```python
# 세그멘테이션 마스크 생성
person_seg_mask = np.all(output==person_seg, axis=-1) 
print(person_seg_mask.shape)
plt.imshow(person_seg_mask, cmap='gray')
plt.show()
```

### 마스크와 원본 이미지 겹쳐 그리기

```python
# 원본 이미지와 마스크를 겹쳐서 출력
# True과 False인 값을 각각 255과 0으로 바꿔줌
person_mask_2d = person_seg_mask.astype(np.uint8) * 255

# 255와 0을 적당한 컬러맵으로 변경
person_mask_3d_color = cv2.applyColorMap(person_mask_2d, cv2.COLORMAP_WINTER) # 3차원 이미지로 변환(BGR)
person_mask_3d_color = cv2.cvtColor(person_mask_3d_color, cv2.COLOR_BGR2RGB)

# 원본 이미지와 마스트를 적당히 합침
# 0.6과 0.4는 두 이미지를 섞는 비율
img_show = cv2.addWeighted(img_orig, 0.6, person_mask_3d_color, 0.4, 0.0)

plt.imshow(img_show)
plt.show()
```

## 배경 블러 처리

```python
# (13,13)은 blurring kernel size
img_blur = cv2.blur(img_orig, (13,13))

plt.imshow(img_blur)
plt.show()
```

```python
# 이미지 색상 채널을 변경(2차원 -> 3차원)
person_mask_3d = cv2.cvtColor(person_mask_2d, cv2.COLOR_GRAY2RGB)

# cv2.bitwise_not(): 이미지가 반전된다. 배경이 0 사람이 255 -> 배경이 255 사람이 0
bg_mask_3d = cv2.bitwise_not(person_mask_3d)

# cv2.bitwise_and()을 사용하여 비트별 and 연산
img_bg_blur = cv2.bitwise_and(img_blur, bg_mask_3d)
plt.imshow(img_bg_blur)
plt.show()
```

## 배경과 사람 합치기

```python
# 세그멘테이션 마스크가 255인 부분만 원본 이미지 값(사람)을 가져오고 
# 아닌 영역은 블러된 이미지 값(배경)사용
img_concat = np.where(person_mask_3d==255, img_orig, img_bg_blur)

plt.imshow(img_concat)
plt.show()
```

# 여러 이미지에 대해 적용

## 블러 처리 함수화
+ 불필요하다고 생각한 비트 마스크를 통한 배경 분리를 제거하였다.
+ np.where에서 블러 처리한 배경 이미지를 넣나, 블러 처리한 전체 이미지를 넣나 차이가 없기 때문이다.

```python
def blur_background(img_path, object_type):
    # 원본 이미지 저장
    img_orig = cv2.imread(img_path)
    # RGB형태로 변경
    img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)
    # 전역 세그멘테이션 모델 사용
    segvalues, output = model.segmentAsPascalvoc(img_path)
    # output이 BGR형식이기때문에 RGB로 변환
    output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
    
    # 클래스 ID별로 지정된 RGB 색상을 담고있는 colormap 생성 코드(이해할 필요 없음. PixelLib에서 그대로 가져온 코드)
    colormap = np.zeros((256, 3), dtype=int)
    ind = np.arange(256, dtype=int)

    for shift in reversed(range(8)):
        for channel in range(3):
            colormap[:, channel] |= ((ind >> channel) & 1) << shift
        ind >>= 3
        
    obj_seg = colormap[object_type] #컬러맵 15(사람)에 해당하는 RGB값 저장
    
    # 세그멘테이션 마스크 생성
    obj_seg_mask = np.all(output==obj_seg, axis=-1) 

    # 전체 이미지 블러 처리
    img_blur = cv2.blur(img_orig, (13,13))

    # True과 False인 값을 각각 255과 0으로 바꿔줌
    obj_mask_2d = obj_seg_mask.astype(np.uint8) * 255
    # 이미지 색상 채널을 변경(2차원 -> 3차원)
    obj_mask_3d = cv2.cvtColor(obj_mask_2d, cv2.COLOR_GRAY2RGB)
    
    # 세그멘테이션 마스크가 255인 부분만 원본 이미지 값(사람)을 가져오고 
    # 아닌 영역은 블러된 이미지 값 전체 사용
    img_concat = np.where(obj_mask_3d==255, img_orig, img_blur)
    
    fig, axes = plt.subplots(1,2, figsize=(16,8))
    axes[0].imshow(img_orig)
    axes[1].imshow(img_concat)
    axes[0].axis('off')
    axes[1].axis('off')
    plt.show()
```

# 크로마키 적용

```python
def chroma_key_background(img_path, background_path, object_type, ):
    # 원본 이미지 저장
    img_orig = cv2.imread(img_path)
    # RGB형태로 변경
    img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)
    
    # 이미지 크기 가져오기
    height = img_orig.shape[0]
    width = img_orig.shape[1]
    
    img_bg = cv2.imread(background_path)
    img_bg = cv2.cvtColor(img_bg, cv2.COLOR_BGR2RGB)
    # 원본 이미지 크기에 맞게 리사이즈
    img_bg = cv2.resize(img_bg, (width, height))
    
    # 전역 세그멘테이션 모델 사용
    segvalues, output = model.segmentAsPascalvoc(img_path)
    # output이 BGR형식이기때문에 RGB로 변환
    output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
    
    # 클래스 ID별로 지정된 RGB 색상을 담고있는 colormap 생성 코드
    colormap = np.zeros((256, 3), dtype=int)
    ind = np.arange(256, dtype=int)
    for shift in reversed(range(8)):
        for channel in range(3):
            colormap[:, channel] |= ((ind >> channel) & 1) << shift
        ind >>= 3
        
    obj_seg = colormap[object_type] # 컬러맵에서 object_type에 해당하는 RGB값 저장
    
    # 세그멘테이션 마스크 생성
    obj_seg_mask = np.all(output==obj_seg, axis=-1)
    
    # True과 False인 값을 각각 255과 0으로 바꿔줌
    obj_mask_2d = obj_seg_mask.astype(np.uint8) * 255
    # 이미지 색상 채널을 변경(2차원 -> 3차원)
    obj_mask_3d = cv2.cvtColor(obj_mask_2d, cv2.COLOR_GRAY2RGB)
    
    img_concat = np.where(obj_mask_3d==255, img_orig, img_bg)
    
    fig, axes = plt.subplots(1, 3, figsize=(16, 8))
    axes[0].imshow(img_orig)
    axes[1].imshow(img_bg)
    axes[2].imshow(img_concat)
    axes[0].axis('off')
    axes[1].axis('off')
    axes[2].axis('off')
    plt.show()
```

```python
img_bg_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/wave.jpeg'
chroma_key_background(img_path2, img_bg_path, 15)
```

## 개선솔루션
왼쪽 이미지의 경우 옷의 경계를 제대로 잡지 못하는 문제가 있었다. 이를 해결하기 위한 방법으로 다음을 제안한다.
+ Active Contour Model (Snake) 적용
	+ 초기 세그멘테이션 마스크의 외곽선을 시작점으로 Active Contour 알고리즘 적용
	+ 이미지의 에지(gradient) 정보를 활용해 마스크 경계를 더 정확한 위치로 이동시킴
	+ 특히 복잡한 윤곽(머리카락, 의류 주름 등)에서 효과적

Active Contour Model(Snake)은 semantic segmentation mask의 오류를 보완할 수 있는 효과적인 솔루션이다. 이 방법은 **에너지 최소화 원리**를 기반으로 윤곽선을 정교하게 조정하여, DeepLab 모델이 생성한 초기 세그멘테이션 마스크의 부정확한 경계를 실제 이미지의 에지 정보에 맞게 미세 조정한다. 특히 머리카락이나 의류 주름과 같은 복잡한 경계, 그리고 얼굴과 목 주변의 부자연스러운 세그멘테이션 결과를 개선하는 데 효과적이다.
